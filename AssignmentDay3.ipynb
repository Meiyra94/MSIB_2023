3. Tambahkan column ‘tahun’ dan ‘bulan’ sesuai dengan keterangan tanggal pada website

```python
df_penumpang_bus_all = pd.DataFrame()

for tahun, semua_bulan in link_penumpang_bus.items():
  for bulan, link in semua_bulan.items():
    try:
      df_penumpang_bus_per_bulan = pd.read_csv(link)
    except UnicodeDecodeError:
      df_penumpang_bus_per_bulan = pd.read_csv(link, encoding='latin1')

    df_penumpang_bus_per_bulan['bulan'] = str(bulan) #menambahkan kolom bulan
    df_penumpang_bus_per_bulan['tahun'] = int(tahun) #menambahkan kolom tahun
    df_penumpang_bus_all = df_penumpang_bus_all.append(df_penumpang_bus_per_bulan).reset_index(drop=True) #memasukkan data per bulan ke dataframe keseluruhan

df_penumpang_bus_all
```

4. Lakukan data manipulation pada jumlah_bus dan jumlah_penumpang supaya berada dalam satu format type. Contoh:
  1. 89.96 => 8996
  2. 9000.0 => 9000
  3. 9000,0 => 9000

karena jumlah_penumpang masi berupa string, ubah ke int

```python
df_penumpang_bus_all.head()
```

```python
df_penumpang_bus_all.dtypes
```

```python
#untuk menampilkan nilai unik dari suatu kolom
df_penumpang_bus_all['jumlah_penumpang'].unique()
```

```python
#Function untuk mengubah tipe data menjadi integer
def ChangeType(x):
  if isinstance(x,str):
    a = x.replace(',','')
    b = float(a)
    c = int(b)
  else:
    c = int(x)
  return c
```

```python
df_penumpang_bus_all['jumlah_penumpang'] = df_penumpang_bus_all['jumlah_penumpang'].fillna(0) #mengisi nilai NaN dengan 0
df_penumpang_bus_all['jumlah_penumpang'] = df_penumpang_bus_all['jumlah_penumpang'].apply(ChangeType) #mengganti tipe data menjadi integer
```

```python
df_penumpang_bus_all.head()
```

```python
#melihat isi kolom 'jumlah_bus'
df_penumpang_bus_all['jumlah_bus'].unique()
```

```python
df_penumpang_bus_all['jumlah_bus'] = df_penumpang_bus_all['jumlah_bus'].fillna(0) #mengisi nilai NaN dengan 0
df_penumpang_bus_all['jumlah_bus'] = df_penumpang_bus_all['jumlah_bus'].apply(ChangeType) #mengganti tipe data menjadi integer
```

5. Lakukan aggregasi sum, mean, dan max dari jumlah_bus dan jumlah_penumpang berdasarkan tahun

```python
df_hasil_agg = df_penumpang_bus_all.groupby('tahun').agg(
    sum_jumlah_bus = ('jumlah_bus','sum'),
    mean_jumlah_bus = ('jumlah_bus','mean'),
    max_jumlah_bus = ('jumlah_bus','max'),
    sum_jumlah_penumpang = ('jumlah_penumpang','sum'),
    mean_jumlah_penumpang = ('jumlah_penumpang','mean'),
    max_jumlah_penumpang = ('jumlah_penumpang','max')
)
```
